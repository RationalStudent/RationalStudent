```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a sample dataset
np.random.seed(0)
data = pd.DataFrame({
    'Category': np.random.choice(['A', 'B', 'C', 'D'], 100),
    'Value': np.random.randn(100),
    'Value2': np.random.randn(100),
    'Time': pd.date_range('2021-01-01', periods=100)
})
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Category', data=data)
plt.title('Bar Chart')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Value'], bins=10, kde=True)
plt.title('Histogram')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Time'], data['Value'])
plt.title('Line Chart')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Value', y='Value2', hue='Category', data=data)
plt.title('Scatter Plot')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Value', data=data)
plt.title('Box Plot')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
data['Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Pie Chart')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
numeric_data = data.select_dtypes(include=[np.number])
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()
```

```python
sns.pairplot(data, hue='Category')
plt.suptitle('Pair Plot', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Category', y='Value', data=data)
plt.title('Violin Plot')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data['Value'], shade=True)
plt.title('Density Plot')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Value'], data['Value2'], s=data['Value'].abs()*100, alpha=0.5)
plt.title('Bubble Chart')
plt.xlabel('Value')
plt.ylabel('Value2')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.fill_between(data['Time'], data['Value'], color='skyblue', alpha=0.4)
plt.plot(data['Time'], data['Value'], color='Slateblue', alpha=0.6)
plt.title('Area Chart')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```
